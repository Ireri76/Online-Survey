```python
# Step 1: Load Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# Step 2: Generate Random Data
# Set a random seed for reproducibility
np.random.seed(42)

# Generate random data
num_records = 100  # Number of records to generate

data = {
    'Age': np.random.randint(18, 70, size=num_records),  # Random ages between 18 and 70
    'Gender': np.random.choice(['Male', 'Female'], size=num_records),  # Random genders
    'Total Income': np.random.randint(3000, 15000, size=num_records),  # Random income between 3000 and 15000
    'Utilities': np.random.randint(100, 500, size=num_records),  # Random utilities expenses
    'Entertainment': np.random.randint(50, 400, size=num_records),  # Random entertainment expenses
    'School Fees': np.random.randint(0, 1000, size=num_records),  # Random school fees
    'Shopping': np.random.randint(20, 800, size=num_records),  # Random shopping expenses
    'Healthcare': np.random.randint(10, 600, size=num_records),  # Random healthcare expenses
}

# Create a DataFrame
df = pd.DataFrame(data)

# Save the DataFrame to a CSV file
df.to_csv('user_data.csv', index=False)
print("Data generated and saved to 'user_data.csv'")
```

```python
# Step 3: Data Visualization in Jupyter Notebook

# Load the data from the CSV file
df = pd.read_csv('user_data.csv')

# Define age categories
bins = [18, 24, 30, 36, 42, 48, 54, 60, 66, 72]  # Adjust the upper limit as necessary
labels = ['18-24', '25-30', '31-36', '37-42', '43-48', '49-54', '55-60', '61-66', '67-72']
df['Age Group'] = pd.cut(df['Age'], bins=bins, labels=labels, right=False)

# Visualization 1: Show the average income by age group
plt.figure(figsize=(12, 6))
sns.barplot(x='Age Group', y='Total Income', data=df, estimator=np.mean)
plt.title('Figure 1: Average Income by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Average Total Income')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('average_income_by_age_group.png')  # Save the figure
plt.show()

# Visualization 2: Show the gender distribution across spending categories
spending_categories = ['Utilities', 'Entertainment', 'School Fees', 'Shopping', 'Healthcare']
plt.figure(figsize=(12, 6))

# Melt the DataFrame for easier plotting
df_melted = df.melt(id_vars=['Gender'], value_vars=spending_categories)

sns.boxplot(x='variable', y='value', hue='Gender', data=df_melted)
plt.title('Figure 2: Gender Distribution Across Spending Categories')
plt.xlabel('Spending Category')
plt.ylabel('Spending Amount')
plt.xticks(rotation=45)
plt.legend(title='Gender')
plt.tight_layout()
plt.savefig('gender_distribution_spending.png')  # Save the figure
plt.show()
```
